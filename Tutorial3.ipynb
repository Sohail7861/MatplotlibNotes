* Colored Scatterplots

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#reading dataset
iris = pd.read_csv('iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.8)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()

#c(color)=>to select plotting attribute
#cmap=>to select different color styles, #alpha=>to increase/decrease opacity, #colorbar=>for labelling 
```

* Plot Size

```python
# Increasing/Decrease the width and size of plot, applicable all type of graphs
plt.figure(figsize=(20,10))  #20->width of graph, 10->height of graph
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.8)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

* Annotations

```python
#Annotations->helps to place text/label on plots
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,'A')
plt.text(2,6,'B')
plt.text(3,7,'C')
plt.text(4,8,'D',fontdict={'color':'red','size':20})
```

```python
#Annotations on dataset
batsman=pd.read_csv('batter.csv')
```

```python
sample=batsman.head(100).sample(25,random_state=5)
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample['avg'],sample['strike_rate'],s=sample['runs'])
for i in range(sample.shape[0]):
    plt.text(sample['avg'].values[i],sample['strike_rate'].values[i],sample['batter'].values[i])
    
```

* Horizontal and Vertical lines

```python
#batsmen with strike_rate>125 and avg_runs>30
plt.figure(figsize=(18,10))
plt.scatter(sample['avg'],sample['strike_rate'],s=sample['runs'])

plt.axhline(125,color='green')
plt.axvline(30,color='red')

for i in range(sample.shape[0]):
    plt.text(sample['avg'].values[i],sample['strike_rate'].values[i],sample['batter'].values[i])
```

* SUB PLOTS

 * Different ways to plot graphs

```python
#method 1
plt.figure(figsize=(15,6))
plt.scatter(batsman['avg'],batsman['strike_rate'],color='green',marker='+')
plt.title("Method 1")
plt.xlabel('AVG')
plt.ylabel('Strike Rate')
plt.show()
```

```python
#method 2
fig, ax=plt.subplots(figsize=(15,6))  #fig=graph, ax=axis
ax.scatter(batsman['avg'],batsman['strike_rate'],color='green',marker='+')
ax.set_title('Method 2')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')

fig.show()

#Method 2 is used when we want to plot multiple graphs
```

```python
plt.subplots(nrows=2,ncols=1) #2 graphs
#plt.subplots(nrows=4,ncols=4) #16 graphs
```

```python
fig, ax=plt.subplots(nrows=2,ncols=1) 
fig #plots graph
```

```python
ax #shows axis
```

```python
#plotting AVGvsSR and AVGvsRuns graphs
fig, ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))  #fig=graph, ax=axis
ax[0].scatter(batsman['avg'],batsman['strike_rate'],color='green',marker='+')
ax[0].set_title('AVG vs SR')
ax[0].set_ylabel('Strike Rate')

ax[1].scatter(batsman['avg'],batsman['runs'],color='red',marker='+')
ax[1].set_title('AVG vs RUNS')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ax[0,0].scatter(batsman['avg'],batsman['strike_rate'],color='yellow')
ax[0,1].scatter(batsman['avg'],batsman['runs'],color='green')
ax[1,0].hist(batsman['avg'],color='red')
ax[1,1].hist(batsman['runs'],color='blue')
```

```python
#Method 3
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batsman['avg'],batsman['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batsman['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batsman['avg'])
```
