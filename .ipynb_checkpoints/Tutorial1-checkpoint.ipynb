* Types of data :
        1. Numerical Data
        2. Categorical Data

* Univariate Analysis : *Focuses on understanding individual variables. 
                        *Graph plotted on single variable or column.

* Bivariate Analysis  : *Examines relationships between two variables. 
                        *Graph plotted on two varibles or two column.

* Multivariate Analysis : *Deals with the interactions and relationships among three or more variables.
                          *Graph plotted on two or more columns.  

```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   #main library
import seaborn as sns

plt.style.use('default')
```

* 2D Line Plot

    *Used in these cases :
        *Bivariate Analysis
        *categorical -> numerical and numerical -> numerical
        *Use case - Time series data

```python
#plotting 2D
price=[2000000,4000000,1000000,500000]    #numerical
year=[2019,2020,2021,2022]                  #categorical
```

```python
plt.plot(year,price)     #X-axis : Categorical data,  Y-axis : Numerical data
```

```python
# from a pandas dataframe,  plotting a 2D Line plot
batsman=pd.read_csv('sharma-kohli.csv')
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
plt.title('RG Sharma vs V Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# colors(hex) 
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.title('RG Sharma vs V Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# line(width and style)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='solid',linewidth=1)
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3)
plt.title('RG Sharma vs V Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# marker(style and size)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='solid',linewidth=1,marker='D',markersize=6)
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=3,marker='o',markersize=10)
plt.title('RG Sharma vs V Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# legend -> location                                     # ---->--<---- # 
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',label='Rohit',linestyle='solid',linewidth=1,marker='D',markersize=6)
plt.plot(batsman['index'],batsman['V Kohli'],color='green',label='Virat',linestyle='dashed',linewidth=3,marker='o',markersize=10)
plt.title('RG Sharma vs V Kohli')                        # ---->--<---- #
plt.xlabel('Year')
plt.ylabel('Runs')
#plt.legend()
plt.legend(loc='upper center')
#Options : best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 
#'center left', 'center right', 'lower center', 'upper center', 'center'
```

```python
# limiting axes
price=[2000000,4000000,1000000,500000,80000000]    
year=[2019,2020,2021,2022,2023] 
plt.plot(year,price)
```

```python
plt.plot(year,price)
plt.ylim(0,6000000)
plt.xlim(2019,2022)
```

```python
# grid
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',label='Rohit',linestyle='solid',linewidth=1,marker='D',markersize=6)
plt.plot(batsman['index'],batsman['V Kohli'],color='green',label='Virat',linestyle='dashed',linewidth=3,marker='o',markersize=10)
plt.title('RG Sharma vs V Kohli')                       
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend(loc='upper center')

plt.grid()     #syntax
```

```python
# show
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',label='Rohit',linestyle='solid',linewidth=1,marker='D',markersize=6)
plt.plot(batsman['index'],batsman['V Kohli'],color='green',label='Virat',linestyle='dashed',linewidth=3,marker='o',markersize=10)
plt.title('RG Sharma vs V Kohli')                       
plt.xlabel('Year')
plt.ylabel('Runs')
plt.legend(loc='upper center')
plt.grid()     

plt.show()   #syntax
```

* Scatter Plots

  *2D Plot is a special case of Scatter Plots. In 2D plots, first
         scatter points are generated and lines are drawn between them.

            *Used in these cases :
                *Bivariate Analysis
                *numerical vs numerical
                *Use case - Finding Correlation

```python
# plt.scatter simple function
x=np.linspace(10,20,50)
y=20*x+3+np.random.randint(1,200,50)
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
data=pd.read_csv('batter.csv')
data=data.head(50)
plt.scatter(data['avg'],data['strike_rate'],color='green',marker='+',)
plt.title('AVG and ST of Top 50 Batsmen')
plt.xlabel('avg')
plt.ylabel('strike rate')
```

```python
# Scatter plot without scatter()
plt.plot(data['avg'],data['strike_rate'],'o') #-> faster techinque than plt.scatter()
```

* Bar Chart

            *Used in these cases :
                *Bivariate Analysis
                *Numerical vs Categorical
                *Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']
plt.bar(colors,children,color='black')
```

```python
# bar chart using data  : same functions used in plot(), scatter() are applicatble in bar()
```

* bar charts two types : Vertical(bar()) and Horizontal(barh())

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
# color and label
dt=pd.read_csv('batsman_season_record.csv')
dt
```

```python
#runs batsman scored all three years
plt.bar(np.arange(dt.shape[0])-0.2, dt['2015'], width=0.2, color='orange')
plt.bar(np.arange(dt.shape[0]), dt['2016'], width=0.2, color='gray')
plt.bar(np.arange(dt.shape[0])+0.2, dt['2017'], width=0.2, color='green')
plt.xticks(np.arange(dt.shape[0]),dt['batsman'])
plt.show()
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['Mahendra Dhoni','Virendhar Sehwag','Sachin Tendulkar','Virat Kohli','Yuvraj Singh']
plt.barh(children,colors)

#overlapping colors on y-axis
```

```python
children = [10,20,40,10,30]
colors = ['Mahendra Dhoni','Virendhar Sehwag','Sachin Tendulkar','Virat Kohli','Yuvraj Singh']
plt.barh(children,colors)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#stacked bar chart
plt.bar(dt['batsman'],dt['2017'],label='2017')
plt.bar(dt['batsman'],dt['2016'],bottom=dt['2017'],label='2016')
plt.bar(dt['batsman'],dt['2015'],bottom=(dt['2017']+dt['2016']),label='2015')
plt.legend()
plt.show()
```
